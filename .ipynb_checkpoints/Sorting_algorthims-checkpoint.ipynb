```python
## Bubble sort

def bubblesort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                ## swap here
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
 
    
arr=[70,20,50,30,90,5,15]
result=bubblesort(arr)
print("Array after applying Bubble sort",result)
## Now, let's step through each iteration of the outer and inner loops:
###In this loop, j is initialized to 0, and it increments in each iteration of the loop, going from 0 to n - i - 2. 
#This range is used to compare adjacent elements in the array and perform swaps as needed to sort the elements.
"""Iteration 1 (Outer Loop - i=0):

Inner Loop (j=0):

Compare arr[0] (5) and arr[1] (2). Since 5 > 2, we swap them.
The array becomes: [2, 5, 9, 3, 4]
Inner Loop (j=1):

Compare arr[1] (5) and arr[2] (9). No swap is needed.
Inner Loop (j=2):

Compare arr[2] (9) and arr[3] (3). Swap them.
The array becomes: [2, 5, 3, 9, 4]
Inner Loop (j=3):

Compare arr[3] (9) and arr[4] (4). Swap them.
The array becomes: [2, 5, 3, 4, 9]
After the first iteration of the outer loop (i=0), the largest element (9) has "bubbled up" to the end of the array.

Iteration 2 (Outer Loop - i=1):

Inner Loop (j=0):

Compare arr[0] (2) and arr[1] (5). No swap is needed.
Inner Loop (j=1):

Compare arr[1] (5) and arr[2] (3). Swap them.
The array becomes: [2, 3, 5, 4, 9]
Inner Loop (j=2):

Compare arr[2] (5) and arr[3] (4). Swap them.
The array becomes: [2, 3, 4, 5, 9]
After the second iteration of the outer loop (i=1), the second largest element (5) has "bubbled up" to the second-to-last position in the array.

Iteration 3 (Outer Loop - i=2):

Inner Loop (j=0):
Compare arr[0] (2) and arr[1] (3). Swap them.
The array becomes: [2, 3, 4, 5, 9]
No swaps are needed in the inner loop during this iteration because the remaining elements are already in sorted order.

Iteration 4 (Outer Loop - i=3):

Inner Loop (j=0):
Compare arr[0] (2) and arr[1] (3). No swap is needed.
The outer loop completes, and the sorting is finished. The final sorted array is:"""
```
